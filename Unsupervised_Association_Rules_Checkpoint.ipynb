```python
import mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
toy_dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],

        ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],

        ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],

        ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],

        ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
len(toy_dataset)
```

```python
te = TransactionEncoder()
transform = te.fit(toy_dataset).transform(toy_dataset)
df = pd.DataFrame(transform, columns=te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori
apriori(df, min_support=0.5)
```

```python
frequent_itemsets = apriori(df, min_support=0.5, use_colnames=True)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7)
```

```python
association_rules(frequent_itemsets,metric="lift",min_threshold=1.01)
```

```python
import pandas as pd
df = pd.read_csv('/content/Market_Basket_Optimisation.csv')
df.head()
```

```python
df.info()
```

```python
# EDA using plotly
import plotly.express as px
from collections import Counter

# Convert the dataset into a list of transactions
transactions = df.apply(lambda row: [item for item in row if pd.notnull(item)], axis=1).tolist()

# Flatten the list of transactions to count item frequencies
all_items = [item for transaction in transactions for item in transaction]
item_counts = Counter(all_items)

# Convert to DataFrame for plotting
item_counts_df = pd.DataFrame(item_counts.items(), columns=["Item", "Count"]).sort_values(by="Count", ascending=False)

# Plot top 20 most frequent items
fig = px.bar(item_counts_df.head(20), x="Item", y="Count", title="Top 20 Most Frequent Items in Transactions")
fig.show()
```

```python
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)
df_encoded
```

```python
# Apply Apriori algorithm
frequent_itemsets = apriori(df_encoded, min_support=0.02, use_colnames=True)
frequent_itemsets
```

```python
# Generate association rules
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.3)
```

```python
association_rules(frequent_itemsets,metric="lift",min_threshold=1.2)
```

# Business Insights Based On Findings From The Data

1) Product Bundling:

If mineral water and spaghetti are frequently bought together, place them near each other in the store or offer bundle discounts.

2) Cross-Promotions:

Offer a discount on chocolate if someone buys eggs and milk, encouraging impulse purchases.

3) Inventory Strategy:

Ensure high-demand items (like mineral water, eggs) are never out of stock.

4) Layout Optimization:

Place co-purchased items in close proximity to reduce friction in purchase decisions.
